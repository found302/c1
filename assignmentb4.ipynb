```python
gender_counts = data['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['lightcoral', 'lightskyblue'])
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['hours-per-week'], bins=20, color='mediumpurple')
plt.title('Distribution of Working Hours per Week')
plt.xlabel('Hours per Week')
plt.ylabel('Frequency')
plt.show()
```

```python
edu_age = data.groupby('education')['age'].mean().sort_values()
plt.figure(figsize=(10, 6))
edu_age.plot(kind='line', marker='o', color='teal')
plt.title('Average Age by Education Level')
plt.ylabel('Average Age')
plt.xlabel('Education Level')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='age', y='hours-per-week', hue='income', data=data, alpha=0.6)
plt.title('Age vs. Working Hours (Colored by Income)')
plt.show()
```

```python
numeric_data = data.select_dtypes(include='number')
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
from wordcloud import WordCloud
text = ' '.join(data['occupation'].dropna().astype(str))
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Occupations')
plt.show()
```

```python
edu_income = pd.crosstab(data['education'], data['income'])

# Sort the index for better readability
edu_income = edu_income.sort_values(by='<=50K', ascending=False)

# Plot stacked bar chart
edu_income.plot(kind='bar', stacked=True, figsize=(12, 6),width=0.8, colormap='viridis')

# Add labels and title
plt.title('Stacked Bar Chart: Education Level vs Income')
plt.xlabel('Education Level')
plt.ylabel('Number of Individuals')

plt.legend(title='Income')
plt.show()
```
